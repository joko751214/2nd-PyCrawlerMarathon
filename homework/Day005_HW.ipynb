```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
import json

r = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
r_dict = json.loads(r.text)
print("1æ¬¡æœƒå›å‚³%dç­†è³‡æ–™" % len(r_dict))
print("æ¯ç­†è³‡æ–™æ¬„ä½åŒ…å«:", r_dict[0].keys())
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
for i in r_dict:
    print("æ¨™é¡Œ:", i['title'])
    print("è²¼æ–‡æ™‚é–“:", i['createdAt'])
    print("ç•™è¨€äººæ•¸:", i['commentCount'])
    print("æŒ‰è®šäººæ•¸:", i['likeCount'], end='\n\n')
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
average_commentCount = 0
average_likeCount = 0
total_count = 0
for i in r_dict:
    average_commentCount += i['commentCount']
    average_likeCount += i['likeCount']
    total_count += 1

print("ç¸½å…±æœ‰%dç¯‡æ–‡ç« " % total_count)
print("å¹³å‡ç•™è¨€äººæ•¸:", round(average_commentCount/total_count))
print("å¹³å‡æŒ‰è®šäººæ•¸:", round(average_likeCount/total_count))
```
