## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
import time

for i in range(20):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
```

```python
import re

html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html5lib')

for d in soup.find('div', class_="part_list_2").find_all('h3'):
    try:
        if re.match('2020/03/17', d.find(class_="date").text).group():
            print(d.find(class_="date").text, d.find_all('a')[-1].text)
    except:
        pass
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import datetime, timedelta

now = datetime.now()
two_hours_ago = now - timedelta(hours=2)
print("現在時間:", now)

for d in soup.find('div', class_="part_list_2").find_all('h3'):
    date_time = datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M")
    try:
        if date_time > two_hours_ago:    
            print(d.find(class_="date").text, d.find_all('a')[-1].text)
        else:
            break
    except:
        pass
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime, timedelta

browser = webdriver.Chrome(executable_path='chromedriver')

now = datetime.now()
three_days_ago = now.date() - timedelta(days=3)
three_days_ago = three_days_ago.strftime("%Y-%m-%d")
url = "https://www.ettoday.net/news/news-list-{}-0.htm".format(three_days_ago)
browser.get(url)
```

```python
import time
import random

for i in range(20):
    print(i)
    time.sleep(random.randint(2,5))
    browser.execute_script("window.scrollTo(0, 20000);")
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html5lib')

three_days_ago_15 = datetime.strptime(three_days_ago, "%Y-%m-%d") + timedelta(hours=15)
three_days_ago_17 = datetime.strptime(three_days_ago, "%Y-%m-%d") + timedelta(hours=17)

for d in soup.find('div', class_="part_list_2").find_all('h3'):
    date_time = datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M")
    try:
        if  date_time < three_days_ago_17:
            if date_time > three_days_ago_15: 
                print(date_time)
                print(d.find(class_="date").text, d.find_all('a')[-1].text)
            else:
                break
    except:
        pass
```
