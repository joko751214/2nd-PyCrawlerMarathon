```python
'''
Your Code
'''
with open("./example/64_72hr_CH.xml", "r", encoding='utf-8') as f:
    xml = f.read()
```

```python
import xmltodict

with open("./example/64_72hr_CH.xml", "r", encoding='utf-8') as f:
    doc = dict(xmltodict.parse(f.read()))
```

### Ans:
* File I/O讀取的資料為一連串的String, 而xmltodict則為dict type。

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
'''
Your Code
'''
location = doc['cwbopendata']['dataset']['locations']['location']
print(len(location))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
'''
Your Code
'''
for local in location:
    print("地點:", local['locationName'], end=' ')
    print("時間:", local['weatherElement'][0]['time'][0]['dataTime'], end=' ')
    print("溫度:", local['weatherElement'][0]['time'][0]['elementValue']["value"])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
'''
Your Code
'''
for weather in location[0]['weatherElement']:
    for time in weather['time']:
        print("時間:", time['dataTime'], "溫度:", time['elementValue']['value'])
```
