### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

r = requests.get("https://www.ettoday.net/news/news-list.htm")
soup = BeautifulSoup(r.text, 'html5lib')

for i in soup.find(class_="part_list_2").find_all('h3'):
#     print(i)
    print(i.find(class_='date').text, i.find('em').text, i.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
data = {}
date, category, title = [], [], []
for i in soup.find(class_="part_list_2").find_all('h3'):
    date.append(i.find(class_='date').text)
    category.append(i.find('em').text)
    title.append(i.find('a').text)
    
data['date'] = date
data['category'] = category
data['title'] = title
print(data)
```

### ③ 哪一個類別的文章最多

```python
import pandas as pd

df = pd.DataFrame(data)
df.category.mode()
```

```python
print(len(df[df.category == '生活']))
print(len(df[df.category == '體育']))
```
