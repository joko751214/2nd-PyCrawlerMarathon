```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
d = {}
for tr in table.find_all('tr')[1:]:
    for i, td in enumerate(tr.find_all('td')):
        if len(tr.find_all('td')) == 5:
            if i == 0:
                title = td.text
                d.setdefault(title, {})
            if i == 2:
                date = td.text
            if i == 3:
                value = td.text
                d[title][date] = value
        else:
            if i == 0:
                date = td.text
            if i == 1:
                value = td.text
                d[title][date] = value
print(d)
```
